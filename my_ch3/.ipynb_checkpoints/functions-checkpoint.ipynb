```python
import numpy as np

def step_function(x):
    y=x>0
    return y.astype(np.int)
```

```python
x=np.array([-1.0,1.0,2.0])
y=x>0
```

```python
print(y)
y.astype(np.int)#astype():引数に希望する型を指定する
```

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.arange(-5.0,5.0,0.1)
y=step_function(x)
plt.xlabel("x")
plt.ylabel("y")
plt.title("step_function")
plt.plot(x,y,linestyle="--",label="step")
plt.ylim(-0.1,1.1)
plt.show()
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
x=np.arange(-5.0,5.0,0.1)
y=sigmoid(x)
plt.plot(x,y)
plt.ylim(-0.1,1.1)
plt.show()
```

```python
def ReLU(x):
    return np.maximum(0,x)
```

```python
x=np.arange(-2.0,2.0,0.05)
y=ReLU(x)
plt.plot(x,y)
plt.show()
```

```python
#3層ニューラルネットワークの実装
X=np.array([1.0, 0.5])
W1=np.array([[0.1, 0.3, 0.5], [0.2, 0.4, 0.6]])
B1=np.array([0.1, 0.2, 0.3])
print(W1.shape)
print(X.shape)
print(B1.shape)
```

```python
A1=np.dot(X,W1)+B1
```

```python
print(A1)
Z1=sigmoid(A1)
print(Z1)
```

```python
W2=np.array([[0.1, 0.4], [0.2, 0.5], [0.3, 0.6]])
B2=np.array([0.1, 0.2])
A2=np.dot(Z1, W2)+B2
Z2=sigmoid(A2)
```

```python
def identify_function(x):
    return x
```

```python
W3=np.array([[0.1, 0.3], [0.2, 0.4]])
B3=np.array([0.1, 0.2])
```

```python
A3=np.dot(Z2, W3)+B3
Y=identify_function(A3)
```
