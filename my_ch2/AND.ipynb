```python
#簡単なパーセプトロンの実装
def AND(x1,x2):
    w1,w2,theta=0.5, 0.5, 0.7
    tmp=x1*w1+x2*w2
    if tmp<=theta:
        return 0
    elif tmp>theta:
        return 1
```

```python
print(AND(0,0))
print(AND(0,1))
print(AND(1,0))
print(AND(1,1))
```

```python
#重みとバイアスによる実装
def AND_bias(x1,x2):
    x=np.array([x1,x2])#入力信号
    w=np.array([0.5,0.5])#重み
    b=-0.7#バイアス
    tmp=np.sum(w*x)+b
    #出力信号
    if tmp>0:
        return 1
    elif tmp<=0:
        return 0
```

```python
print(AND_bias(0,0))
print(AND_bias(0,1))
print(AND_bias(1,0))
print(AND_bias(1,1))
```
